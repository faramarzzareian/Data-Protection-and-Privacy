```python
import csv
from faker import Faker
import datetime

def datagenerate(records, headers):
    fake = Faker('en_US')
    fake1 = Faker('en_GB')    
    with open("test data.csv", 'wt') as csvFile:
        writer = csv.DictWriter(csvFile, fieldnames=headers)
        writer.writeheader()
        for i in range(records):
            full_name = fake.name()
            FLname = full_name.split(" ")
            Fname = FLname[0]
            Lname = FLname[1]
            domain_name = "@testDomain.com"
            userId = Fname +"."+ Lname + domain_name

            writer.writerow({
                     
                    "Name": fake.name(),
                    
                    "Birth_Date" : fake.date(pattern="%d-%m-%Y", end_datetime=datetime.date(2000, 1,1)),
                    "Phone" : fake1.phone_number(),
                    "Email": fake.email(),
                    "Address" : fake.address(),
                    "Zip" : fake.zipcode(),
                    "City" : fake.city(),
                    "State" : fake.state(),
                    "Country" : fake.country(),
                    "Year":fake.year(),
                    "Lon":fake.longitude(),
                    "Lat":fake.latitude(),})
if __name__ == '__main__':
    records = 20
    headers = [  "Name",  "Birth_Date", "Phone", "Email","Address", "Zip", "City","State", "Country", "Year","Lon","Lat" ]
    datagenerate(records, headers)
    print("CSV is done!")
```
